## Tuples

   - Tuples are ordered sequence of mixed data types
   - Tuples are written in comma separated elements within parenthesis

```python
t = (1,2,3,4)
```

```python
type(t)
```

```python
t = ('hello', 5, 5.5)
t
```

```python
len(t)
```

```python
t1 = ('hello', 5, 5.5, [2,3,4])
```

```python
var = t1[3]
type(var)
```

```python
var[1]
```

```python
t1[3][1]
```

```python
len(t1)
```

    Tuples can be defined in various ways

```python
t1 = 1,2,3,4,5
print(t1)
print(type(t1))
```

```python
t1 = 1.2,
type(t1)
```

```python
len(t1)
```

```python
t2 = (2,)
```

```python
t3 = 2,
```

```python
type(t3)
```

```python
#single value tuple

t2 = 1,
print(t2)
```

```python
t3 = (1,)
print(t3)
```

```python
#This is not a tuple

t4 = 1
t5 = (1)
```
