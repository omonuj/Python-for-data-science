### Loops

#### Check whether a number is even or odd

```python
x = 101

if x % 2 == 0:
    print(x, "is an Even number")
else:
    print(x, "is an Odd number")
```

```python
x = int(input("Enter your age: "))

if x % 2 == 0:
    print(x, "is an Even number")
else:
    print(x, "is an Odd number")
```
