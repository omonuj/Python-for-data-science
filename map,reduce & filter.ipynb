## Map, Reduce & Filter

#### Area of a circle

```python
import math

def area(r):
    return math.pi*(r**2)

radii = [1,2,3,4,5]

areas = []

for r in radii:
    a = area(r)
    areas.append(a)
```

```python
print(areas)
```

#### map(f,iterable object)

```python
map(area, radii)
```

```python
list(map(area, radii))
```
