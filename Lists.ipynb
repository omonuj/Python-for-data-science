```python
list1 = ["India", 5, "Delhi"]
```

```python
#Nested list - List within a List

list2 = ["India", "Delhi", 5, 10, ["Rajkot", "Chennai"]]
len(list2)
```

```python
list2
```

```python
list2[-1][1]
```
