#### Mutability

```python
list1 = ["Maths", "Physics", "Chemistry", "Biology"]
```

```python
list1[3] = 'Computer Science'
```

```python
list1
```

```python
#extend() function --> It extends the list

list1 = [1,2,3,4]
list1.extend([5,6])
len(list1)
```

```python
list1 = [1,2,3,4]
list1.append([5,6])
len(list1)
```

```python
#append() function

list1 = [1,2,3,4]
list1.append([5,6])
```

```python
#del command

list1 = [1,2,3,4]

del list1[2]
list1
```

```python
#pop() - Deletes the object based on the index value
# Default --> Removes the last element

list1 = [11,22,33,44]

list1.pop(2)
list1
```

```python
list1 = [11,22,33,44]

list1.remove(22)
list1
```

```python
#pop() - By default, it deletes the last object from the list
#pop takes index as an input

list1 = [1,2,3,4]

list1.pop(2)
list1
```

```python
#remove() - takes value as an input

list1 = [1,2,3,4]

list1.remove(3)
```

#### Difference between sort & sorted

```python
A = ["Oranges", "Strawberry", "Mango"]
B = A.sort()
print(A)
print(B)
```

```python
A = ["Oranges", "Strawberry", "Mango"]
A.sort()
```

```python
A = ["Oranges", "Strawberry", "Mango"]
B = sorted(A)
print(A)
print(B)
```

#### shallow copy

```python
A = ["Oranges", "Strawberry", "Mango"]
B = A
print(B)
print(A)
```
